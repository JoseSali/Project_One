```python
import pandas as pd
import numpy as np

# Define a function to remove non-numeric characters
def remove_non_numeric(value):
    return ''.join(filter(str.isdigit, str(value)))

# Convert columns to float types, handling periods ('.') gracefully
countries = ['Germany', 'Netherlands', 'Switzerland', 'United Kingdom', 'United States of America']
for country in countries:
    data_2010_summary.loc[:, country] = data_2010_summary[country].apply(remove_non_numeric).astype(float)

# Replace any resulting NaN values with 0
data_2010_summary.loc[:, countries] = data_2010_summary[countries].fillna(0)
```
