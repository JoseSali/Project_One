```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Tax_Data = 'Resources/TAX_DATA.csv'
tax_data_df = pd.read_csv(Tax_Data)
tax_data_df.head()
```

```python

# Sample DataFrame
data = {'Countruy': ['Germany', 'Netherlands', 'Switzerland', 'United Kingdom', 'United States', 'France', 'Italy', 'Spain']}

# List of countries to filter
countries_to_keep = ['Germany', 'Netherlands', 'Switzerland', 'United Kingdom', 'United States']

# Filtering the DataFrame
filtered_df = tax_data_df[tax_data_df['Country'].isin(countries_to_keep)]

filtered_df.head()
```

```python
tax_copy_1 = filtered_df.copy()

# Total Tax
total_income_tax_df = tax_copy_1.loc[(tax_copy_1['Year'] >= 2009) & 
                              (tax_copy_1['Year'] <= 2019) & 
                              (tax_copy_1['Indicator'] == 'Tax revenue as % of GDP') & 
                              (tax_copy_1['TAX'] == '1000')&
                              (tax_copy_1['Level of government'] == 'Total')]

# Payroll Tax (Taxes on payroll and workforce)
payroll_df = tax_copy_1.loc[(tax_copy_1['Year'] >= 2009) & 
                             (tax_copy_1['Year'] <= 2019) & 
                             (tax_copy_1['Indicator'] == 'Tax revenue as % of GDP') & 
                             (tax_copy_1['TAX'] == '3000')&
                             (tax_copy_1['Level of government'] == 'Total')]

# Taxes on Income, Profits, and Capital Gains of Individuals
individual_tax_df = tax_copy_1.loc[(tax_copy_1['Year'] >= 2009) & 
                                    (tax_copy_1['Year'] <= 2019) & 
                                    (tax_copy_1['Indicator'] == 'Tax revenue as % of GDP') & 
                                    (tax_copy_1['TAX'] == '1110')&
                                    (tax_copy_1['Level of government'] == 'Total')]
# Corporate Tax (Taxes on income, profits, and capital gains of corporates)
corporate_tax_df = tax_copy_1.loc[(tax_copy_1['Year'] >= 2009) & 
                                  (tax_copy_1['Year'] <= 2019) & 
                                  (tax_copy_1['Indicator'] == 'Tax revenue as % of GDP') & 
                                  (tax_copy_1['TAX'] == '1200')&
                                  (tax_copy_1['Level of government'] == 'Total')]

```

```python
#Averages
total_tax_ave = total_income_tax_df.groupby('Country')['Value'].mean()
payroll_ave = payroll_df.groupby('Country')['Value'].mean()
corporate_ave = corporate_tax_df.groupby('Country')['Value'].mean()
```

```python
import matplotlib.pyplot as plt

# Plotting Total Tax Change Over Time
plt.figure(figsize=(10, 6))
for country in total_income_tax_df['Country'].unique():
    country_data = total_income_tax_df[total_income_tax_df['Country'] == country]
    plt.plot(country_data['Year'], country_data['Value'], label=country)
plt.xlabel('Year')
plt.ylabel('Tax Revenue as % of GDP')
plt.title('Total Tax Revenue Change Over Time')
plt.legend()
plt.grid(True)
plt.xlim(2009, 2019)  # Setting x-axis limits
plt.tight_layout()
plt.show()


# Plotting Taxes on Income, Profits, and Capital Gains of Individuals Change Over Time
plt.figure(figsize=(10, 6))
for country in individual_tax_df['Country'].unique():
    country_data = individual_tax_df[individual_tax_df['Country'] == country]
    plt.plot(country_data['Year'], country_data['Value'], label=country)
plt.xlabel('Year')
plt.ylabel('Tax Revenue as % of GDP')
plt.title('Taxes on Income, Profits, and Capital Gains of Individuals Change Over Time')
plt.legend()
plt.grid(True)
plt.xlim(2009, 2019)  # Setting x-axis limits
plt.tight_layout()
plt.show()

# Plotting Corporate Tax Change Over Time
plt.figure(figsize=(10, 6))
for country in corporate_tax_df['Country'].unique():
    country_data = corporate_tax_df[corporate_tax_df['Country'] == country]
    plt.plot(country_data['Year'], country_data['Value'], label=country)
plt.xlabel('Year')
plt.ylabel('Tax Revenue as % of GDP')
plt.title('Corporate Tax Revenue Change Over Time')
plt.legend()
plt.grid(True)
plt.xlim(2009, 2019)  # Setting x-axis limits
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Aggregate data for each tax category
total_tax_sum = total_income_tax_df.groupby('Year')['Value'].sum()
payroll_tax_sum = payroll_df.groupby('Year')['Value'].sum()
individual_tax_sum = individual_tax_df.groupby('Year')['Value'].sum()
corporate_tax_sum = corporate_tax_df.groupby('Year')['Value'].sum()

# Plotting Total Tax Pie Chart
plt.figure(figsize=(10, 6))
plt.pie(total_tax_sum, labels=total_tax_sum.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Tax Revenue Distribution Over Time')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Plotting Payroll Tax Pie Chart
plt.figure(figsize=(10, 6))
plt.pie(payroll_tax_sum, labels=payroll_tax_sum.index, autopct='%1.1f%%', startangle=140)
plt.title('Payroll Tax Revenue Distribution Over Time')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Plotting Taxes on Income, Profits, and Capital Gains of Individuals Pie Chart
plt.figure(figsize=(10, 6))
plt.pie(individual_tax_sum, labels=individual_tax_sum.index, autopct='%1.1f%%', startangle=140)
plt.title('Taxes on Income, Profits, and Capital Gains of Individuals Distribution Over Time')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Plotting Corporate Tax Pie Chart
plt.figure(figsize=(10, 6))
plt.pie(corporate_tax_sum, labels=corporate_tax_sum.index, autopct='%1.1f%%', startangle=140)
plt.title('Corporate Tax Revenue Distribution Over Time')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
