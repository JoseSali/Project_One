```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib as mlp
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from matplotlib.transforms import offset_copy
```

```python
#import happiness data
data_c = Path('World Happiness Report.csv')
happy_df = pd.read_csv(data_c)
```

```python
#remove unneeded years
unneeded_years = years_to_exclude =[2005, 2020, 2021, 2022]
happy_df = happy_df[~happy_df['Year'].isin(unneeded_years)]

#remove unneeded countries
happy_df = happy_df[happy_df['Country Name'].isin(['Switzerland', 'Netherlands', 'United Kingdom', 'United States', 'Germany'])]

#rename columns for clarity
happy_df = happy_df.rename(columns={'Log GDP Per Capita': 'GDP'})
```

```python
#filter happiness data into separate data frames

switzerland = happy_df[happy_df['Country Name'].isin(['Switzerland'])]
netherlands = happy_df[happy_df['Country Name'].isin([ 'Netherlands'])]
uk = happy_df[happy_df['Country Name'].isin([ 'United Kingdom'])]
us = happy_df[happy_df['Country Name'].isin(['United States'])]
germany = happy_df[happy_df['Country Name'].isin(['Germany'])]
```

```python

#This graph illustrates the annual happiness rankings worldwide, derived from life evaluations gathered 
#through the Gallup World Poll. Respondents are prompted to envision a ladder, where 10 represents the best possible
#life and 0 denotes the worst. They then rate their current lives on a scale from 0 to 10. The graph visually represents
#the distribution of these life evaluations, providing insight into the overall happiness levels across different regions or time periods.

#plot happiness by country over time
#set figure size
plt.figure(figsize=(10,6))

#plot each country
plt.plot(switzerland['Year'], switzerland['Life Ladder'], label='Switzerland', linewidth=3, marker = 'h', color = 'cyan')
plt.plot(netherlands['Year'],netherlands['Life Ladder'], label='Netherlands',linewidth=3, marker = 'h', color = 'magenta')
plt.plot(uk['Year'], uk['Life Ladder'], label='United Kingdom',linewidth=3, marker = 'h', color='yellow')
plt.plot(germany['Year'], germany['Life Ladder'], label='Germany',linewidth=3, marker = 'h', color = 'red')
plt.plot(us['Year'], us['Life Ladder'], label='United States',linewidth=3, marker = 'h', color = 'purple')

#set ticks and labels
plt.yticks([6,7,8.5])
plt.xlabel('Year')
plt.ylabel('Ranking')
plt.grid(True)
plt.title('Global Happiness Rankings: Life Evaluations on the Cantril Ladder')
plt.legend()

#show plot
plt.show()

```
