# Hypertension Death Rate by Country

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
complete_death_data = "death-rates-from-high-blood-pressure.csv"

complete_death_data = pd.read_csv("death-rates-from-high-blood-pressure.csv")

# Define a list of specific countries you want to include
countries_of_interest = ["Germany", "Netherlands", "Switzerland", "United Kingdom", "United States"]

# Filter the DataFrame to only include rows where the country is in the list of countries of interest
filtered_data = complete_death_data[complete_death_data['Entity'].isin(countries_of_interest)]
filtered_data.rename(columns = {'Deaths that are from all causes attributed to high systolic blood pressure per 100,000 people, in both sexes aged age-standardized':'Deaths'}, inplace = True) 

# Display the first few rows of each group
filtered_data
```

```python
# Assuming 'filtered_data' is a pandas DataFrame containing data for multiple countries
# Extracting unique country names from the DataFrame
country_names = filtered_data["Entity"].unique()

# Plotting each country's data
for country in country_names:
    # Filter data for the current country
    country_data = filtered_data[filtered_data['Entity'] == country]
    
    # Plotting the line chart for the current country
    plt.plot(country_data['Year'], country_data['Deaths'], label=country)

# Adding title and labels
plt.title("Blood Pressure Deaths by Country 1990-2019")
plt.xlabel('Years')
plt.ylabel('Deaths per 100,000')

# Adding a legend
plt.legend()
plt.grid(True)
# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'filtered_data' is a pandas DataFrame containing data for multiple countries
# Extracting unique country names from the DataFrame
country_names = filtered_data["Entity"].unique()

# Filtering data for the specified years
filtered_data = filtered_data[(filtered_data['Year'] >= 2009) & (filtered_data['Year'] <= 2019)]

# Initialize an array to store the cumulative sum of death counts for each year
bottom = np.zeros(len(filtered_data["Year"].unique()))

# Plotting each country's data
for country in country_names:
    # Filter data for the current country
    country_data = filtered_data[filtered_data['Entity'] == country]
    
    # Plotting the bar chart for the current country
    plt.bar(country_data['Year'], country_data['Deaths'], label=country, bottom=bottom)
    
    # Update the cumulative sum of death counts for the next country
    bottom += country_data["Deaths"].to_numpy()

# Adding title and labels
plt.title("Blood Pressure Deaths by Country 2009-2019")
plt.xlabel('Years')
plt.ylabel('Deaths per 100,000')

# Adding a legend
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Display the plot
plt.grid(True)
plt.show()
```
